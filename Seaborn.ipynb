```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'], kde=True)
```

```python
sns.displot(tips['total_bill'], kde=False)
```

```python
sns.jointplot(x='total_bill', y='tip', data = tips)
```

```python
sns.pairplot(tips, hue='sex')
```

```python
sns.factorplot(x = 'day', y = 'total_bill', data = tips, kde = True)
```

#Matrix Plots

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tips.head()
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc, annot=True, cmap = 'coolwarm')
```

```python
tc
```

```python
fp = flights.pivot_table(index = 'month', columns='year', values= 'passengers')
```

```python
sns.heatmap(fp, cmap = 'magma', linecolor='white', linewidths=3)
```

```python
sns.clustermap(fp, cmap = 'magma', linecolor='white', linewidths=3)
```

#Grids

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data = tips, col = 'time', row = 'smoker')
g.map(sns.distplot, 'total_bill')
```

#Regression Plots

```python
sns.lmplot(x = 'total_bill', y = 'tip', data = tips, hue = 'sex', markers = ['o', 'v'], scatter_kws={'s':100})
```
